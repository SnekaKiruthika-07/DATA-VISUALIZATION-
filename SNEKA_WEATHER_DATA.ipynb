```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to read and analyze weather data
def analyze_weather_data(filename):
    # Read the dataset from the CSV file
    data = pd.read_csv(filename)

    # Assuming the dataset contains a 'Date' and 'Temperature' column
    data['Date'] = pd.to_datetime(data['Date'])

    # Plotting temperature trends
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Average humidity (%)'], marker='o', linestyle='-', color='b')
    plt.title('Temperature Trends')
    plt.xlabel('Date')
    plt.ylabel('Temperature (°C)')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Prompt the user for the CSV file name containing the weather data
    filename = '/content/climate_data.csv'
    try:
        analyze_weather_data(filename)
    except FileNotFoundError:
        print("File not found. Please check the file path.")
    except Exception as e:
        print("An error occurred:", e)
```
